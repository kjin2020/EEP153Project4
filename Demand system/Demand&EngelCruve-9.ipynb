```python
import numpy as np
from math import log
import pandas as pd
import matplotlib.pyplot as plt
```

```python
foods = "Haricot Beans"
ref_price = r.prices.sel(t=t,m=m,u='Kg')
def x(n): #Price of good,n is the price scale.
    return n*ref_price.loc[foods].values
def y(h): #Quantity demand of good,h is the price scale, h=n.
    return r.demands(medianx, p={foods:h})[foods]/ref_price.loc[foods].values
```

```python
#Convert price scale to real price.
Prices = np.array([x(0.005),x(0.006),x(0.007),x(0.008),x(0.009),x(0.01),x(0.02),x(0.03),x(0.04),x(0.05),x(0.06),x(0.07),x(0.08),x(0.09),x(0.1),x(0.15),x(0.2),x(0.225),x(0.25),x(0.3),x(0.325),x(0.35),x(0.4),x(0.45),x(0.5),x(0.55),x(0.6),x(0.65),x(0.7),x(0.75),x(0.8),x(0.85),x(0.9),x(0.95),x(1),x(1.25),x(1.5),x(2),x(2.25),x(2.5),x(3),x(3.25),x(3.5),x(4),x(4.5),x(5),x(5.5),x(6)])
#Quantity demand with given price.
D1 = np.array([y(0.005),y(0.006),y(0.007),y(0.008),y(0.009),y(0.01),y(0.02),y(0.03),y(0.04),y(0.05),y(0.06),y(0.07),y(0.08),y(0.09),y(0.1),y(0.15),y(0.2),y(0.225),y(0.25),y(0.3),y(0.325),y(0.35),y(0.4),y(0.45),y(0.5),y(0.55),y(0.6),y(0.65),y(0.7),y(0.75),y(0.8),y(0.85),y(0.9),y(0.95),y(1),y(1.25),y(1.5),y(2),y(2.25),y(2.5),y(3),y(3.25), y(3.5),y(4),y(4.5),y(5),y(5.5),y(6)])
```

```python
def nqd(NQD):
    NP = np.interp(NQD,D1[::-1],Prices[::-1])
    plt.plot(D1,Prices,color="navy",alpha = 0.5)
    plt.annotate("S1", xy = (5.5*y(1),1.01*x(1)))
    plt.annotate("D", xy = (5.5*y(1),x(0.0)))
    plt.axhline(y = x(1), color = "red", alpha = 0.5)
    plt.axvline(x = y(1), color = "black", alpha = 0.5, linestyle = "--")
    plt.axvline(x = NQD, color = "black", alpha = 0.5, linestyle = "--")
    plt.axhline(y = NP, color = "Red", alpha = 0.5)
    plt.annotate("S2", xy = (5.5*y(1),1.01*NP))
    plt.xlabel("%s in Kgs" % foods)
    plt.ylabel("Price")
    plt.xlim([0,6*y(1)])
    plt.show()
    C = round((x(1)-NP)*NQD,2)
    Tax = round(x(1)-NP,2)
    E = NP*NQD
    if NQD > y(1):
        print("Subsidy: ", format(Tax))
    else:
        print("Tax: ", format(-Tax))
    print("Consumer Expenditure: ", format(E))
interactive(lambda NQD: nqd(NQD=NQD), NQD = BoundedFloatText(value=y(1), min = 0, max = 600*y(1), step=0.00001))
```

```python
hh = len(np.array(r.coords["j"]))
e,g,f,h = min25x/medianx,max75x/medianx,medianx/min25x,max75x/min25x
```

```python
#Estimating DWL.
def nqd2(SQD,Q25):
    NP = np.interp(SQD,D1[::-1],Prices[::-1])
    D2 = list(map(lambda x:x*e, D1))
    D3 = list(map(lambda x:x*g, D1))
    plt.plot(D1,Prices,D2,Prices,D3,Prices,color="navy",alpha = 0.5)
    plt.annotate("S1", xy = (5.5*y(1),1.01*x(1)))
    plt.annotate("D1", xy = (y(0.15),x(1)*0.2))
    plt.annotate("D2", xy = (e*y(0.15),x(1)*0.2)) #Demand curve for 25 percentile income household.
    plt.annotate("D3", xy = (g*y(0.15),x(1)*0.2)) #Demand curve for 75 percentile income household.
    P251 = np.interp(Q25,D1[::-1],Prices[::-1])
    NP2 = np.interp(SQD,D2[::-1],Prices[::-1])
    P252 = np.interp(Q25,D2[::-1],Prices[::-1])
    NP3 = np.interp(SQD,D3[::-1],Prices[::-1])
    P253 = np.interp(Q25,D3[::-1],Prices[::-1])
    plt.axvline(x = y(1), color = "black",linestyle = "-")
    plt.axhline(y = x(1), color = "red", alpha = 0.5)
    plt.axvline(x = e*SQD, color = "black",linestyle = "--")
    plt.axvline(x = g*SQD, color = "black",linestyle = "--")
    plt.axvline(x = SQD, color = "black", linestyle = "--")
    plt.axhline(y = P252, color = "Red", alpha = 0.5)
    plt.annotate("S2", xy = (5.5*y(1),1.01*NP))
    plt.axvline(x = f*Q25, color = "darkgreen",linestyle = "--")
    plt.axvline(x = h*Q25, color = "darkgreen",linestyle = "--")
    plt.axvline(x = Q25, color = "darkgreen",linestyle = "--")
    plt.axhline(y = P252, color = "Red", alpha = 0.5)
    plt.annotate("S25", xy = (5.5*y(1),1.01*P252))
    plt.xlabel("%s in Kgs" % foods)
    plt.ylabel("Price")
    plt.xlim([0,4*g*y(1)])
    plt.show()
    Tax2 = round(x(1) - P252,2)
    DWLS1D2 = (x(1) - np.interp(y(1),D2[::-1],Prices[::-1]))*(y(1)-e*y(1))/2
    DWLS1D3 = (np.interp(y(1),D3[::-1],Prices[::-1]) - x(1))*(g*y(1)-y(1))/2
    DWLS2D2 = (NP2 - NP)*(e*SQD-SQD)/2
    DWLS2D3 = (NP3 - NP)*(g*SQD-SQD)/2
    DWLSMCD2 = (P252 - P252)*(Q25-SQD)/2
    DWLSMCD3 = (P253 - P252)*(h*SQD-SQD)/2
    DWLSMCD1 = (P251 - P252)*(f*SQD-SQD)/2
    if Q25 > e*y(1): 
        print("Subsidy: ", format(Tax2))
    else:
        print("Tax: ", format(-Tax2))
    print("DWLP1: ",format(DWLS1D2+DWLS1D3))
    print("DWLP2: ",format(DWLS2D2+DWLS2D3))
    print("DWLP3: ",format(DWLSMCD2+DWLS2D3+DWLSMCD1))
interactive(lambda SQD,Q25: nqd2(SQD=SQD,Q25=Q25),
            Q25 = BoundedFloatText(value=0, min = 0, max = maxx/0.001*y(1), step=0.001),
            SQD = BoundedFloatText(value=y(1), min = 0, max = maxx/0.001*y(1), step=0.001))
```

```python
#Construct a Engel Curve
def nqd3(NQD):
    BC1 = np.linspace(0,y(1),100)
    BC1_1 = np.linspace(y(1),y(1),100)
    BC2 = np.linspace(0,g*y(1),100)
    BC2_1 = np.linspace(g*y(1),g*y(1),100)
    BC3 = np.linspace(0,e*y(1),100)
    BC3_1 = np.linspace(e*y(1),e*y(1),100)
    BC4 = np.linspace(0,NQD,100)
    BC4_1 = np.linspace(NQD,NQD,100)
    Q1 = np.linspace(medianx,0,100)
    Q2 = np.linspace(g*medianx,0,100)
    Q3 = np.linspace(e*medianx,0,100)
    Q4 = np.linspace(np.array((medianx/y(1)) * NQD),0,100)
    P1 = np.linspace(medianx,medianx,100)
    P2 = np.linspace(g*medianx,g*medianx,100)
    P3 = np.linspace(e*medianx,e*medianx,100)
    EC = np.linspace(0,1.5*g*y(1),100)
    P4 = np.linspace(0,1.5*g*medianx,100)
    NB = np.linspace(np.array((medianx/y(1)) * NQD),np.array((medianx/y(1)) * NQD),100)
    fig, ax = plt.subplots()
    ax.plot([e*y(1),y(1),g*y(1),NQD],[e*medianx,medianx,g*medianx,np.array((medianx/y(1)) * NQD)],ls="",marker="o")
    plt.annotate("Engel Curve for "+ foods, xy = (1.6*y(1),2.1*medianx))
    plt.xlim([0,6*y(1)])
    plt.ylim([0,1000])
    plt.xlabel("%s in Kgs" % foods)
    plt.ylabel("Budget")
    plt.plot(BC1,P1,"black",BC2,P2,"black",BC3,P3,"black",BC4,NB,"black",BC1_1,Q1,"black",BC2_1,Q2,"black",BC3_1,Q3,"black",BC4_1,Q4,"black",linestyle="--")
    plt.plot(EC,P4,"darkgreen")
    plt.show()
    Budget = int(np.array((medianx/y(1)) * NQD))
    print("HHBudget:" , format(Budget))
interactive(lambda NQD: nqd3(NQD=NQD),
            NQD = BoundedFloatText(value=y(1), min = 0, max = maxx/0.001*y(1), step=0.001))
```
