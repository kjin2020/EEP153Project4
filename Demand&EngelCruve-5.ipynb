```python
t = r.firstround  # First round
m = r.coords['m'][0] # First market

x = r.get_predicted_expenditures().sum('i')
medianx = x.where(x>0).sel(t=t,m=m).median('j').values # Budget (median household)
max75x = x.where(x>0).sel(t=t,m=m).quantile(0.75,'j').values
min25x = x.where(x>0).sel(t=t,m=m).quantile(0.25,'j').values
```

```python
import numpy as np
from math import log
import pandas as pd
import matplotlib.pyplot as plt
```

```python
foods = "Eggs"
ref_price = r.prices.sel(t=t,m=m,u='Kg')
def x(n): #Price of good,n is the price scale.
    return n*ref_price.loc[foods].values
def y(h): #Quantity demand of good,h is the price scale, h=n.
    return r.demands(medianx, p={foods:h})[foods]/ref_price.loc[foods].values
```

```python
#Convert price scale to real price.
Prices = np.array([x(0.005),x(0.006),x(0.007),x(0.008),x(0.009),x(0.01),x(0.02),x(0.03),x(0.04),x(0.05),x(0.06),x(0.07),x(0.08),x(0.09),x(0.1),x(0.15),x(0.2),x(0.25),x(0.3),x(0.35),x(0.4),x(0.45),x(0.5),x(0.55),x(0.6),x(0.65),x(0.7),x(0.75),x(0.8),x(0.85),x(0.9),x(0.95),x(1),x(1.5),x(2),x(2.5),x(3),x(3.5),x(4),x(4.5),x(5)])
#Quantity demand with given price.
D1 = np.array([y(0.005),y(0.006),y(0.007),y(0.008),y(0.009),y(0.01),y(0.02),y(0.03),y(0.04),y(0.05),y(0.06),y(0.07),y(0.08),y(0.09),y(0.1),y(0.15),y(0.2),y(0.25),y(0.3),y(0.35),y(0.4),y(0.45),y(0.5),y(0.55),y(0.6),y(0.65),y(0.7),y(0.75),y(0.8),y(0.85),y(0.9),y(0.95),y(1),y(1.5),y(2),y(2.5),y(3),y(3.5),y(4),y(4.5),y(5)])
```

```python
def nqd(NQD):
    NP = np.interp(NQD,D1[::-1],Prices[::-1])
    plt.plot(D1,Prices,color="navy",alpha = 0.5)
    plt.annotate("S1", xy = (5.5*y(1),1.01*x(1)))
    plt.annotate("D", xy = (5.5*y(1),x(0.0)))
    plt.axhline(y = x(1), color = "red", alpha = 0.5)
    plt.axvline(x = y(1), color = "black", alpha = 0.5, linestyle = "--")
    plt.axvline(x = NQD, color = "black", alpha = 0.5, linestyle = "--")
    plt.axhline(y = NP, color = "Red", alpha = 0.5)
    plt.annotate("S2", xy = (5.5*y(1),1.01*NP))
    plt.xlabel("%s in Kgs" % foods)
    plt.ylabel("Price")
    plt.xlim([0,6*y(1)])
    plt.show()
    C = round((x(1)-NP)*NQD,2)
    Tax = round(x(1)-NP,2)
    if NQD > y(1):
        print("Subsidy: ", format(Tax))
    else:
        print("Tax: ", format(-Tax))
interactive(lambda NQD: nqd(NQD=NQD), NQD = BoundedFloatText(value=0, min = 0, max = 600*y(1), step=0.00001))
```

```python
hh = len(np.array(r.coords["j"]))
e,g,f,h = min25x/medianx,max75x/medianx,medianx/min25x,max75x/min25x
```

```python
#Estimating DWL.
def nqd2(NQD,SQD):
    NP = np.interp(NQD,D1[::-1],Prices[::-1])
    D2 = list(map(lambda x:x*e, D1))
    D3 = list(map(lambda x:x*g, D1))
    plt.plot(D1,Prices,D2,Prices,D3,Prices,color="navy",alpha = 0.5)
    plt.annotate("S1", xy = (5.5*y(1),1.01*x(1)))
    plt.annotate("D1", xy = (y(0.15),x(1)*0.2))
    plt.annotate("D2", xy = (e*y(0.15),x(1)*0.2)) #Demand curve for 25 percentile income household.
    plt.annotate("D3", xy = (g*y(0.15),x(1)*0.2)) #Demand curve for 75 percentile income household.
    SP = np.interp(SQD,D1[::-1],Prices[::-1])
    NP2 = np.interp(NQD,D2[::-1],Prices[::-1])
    SP2 = np.interp(SQD,D2[::-1],Prices[::-1])
    NP3 = np.interp(NQD,D3[::-1],Prices[::-1])
    SP3 = np.interp(SQD,D3[::-1],Prices[::-1])
    plt.axvline(x = y(1), color = "black",linestyle = "--")
    plt.axhline(y = x(1), color = "red", alpha = 0.5)
    plt.axvline(x = e*NQD, color = "black",alpha = 0.5,linestyle = "--")
    plt.axvline(x = g*NQD, color = "black",alpha = 0.5,linestyle = "--")
    plt.axvline(x = NQD, color = "black", alpha = 0.5, linestyle = "--")
    plt.axhline(y = NP, color = "Red", alpha = 0.5)
    plt.annotate("S2", xy = (5.5*y(1),1.01*NP))
    plt.axvline(x = e*SQD, color = "darkgreen",alpha = 0.5,linestyle = "--")
    plt.axvline(x = g*SQD, color = "darkgreen",alpha = 0.5,linestyle = "--")
    plt.axvline(x = SQD, color = "darkgreen", alpha = 0.5, linestyle = "--")
    plt.axhline(y = SP, color = "Red", alpha = 0.5)
    plt.annotate("SMC", xy = (5.5*y(1),1.01*SP))
    plt.xlabel("%s in Kgs" % foods)
    plt.ylabel("Price")
    plt.xlim([0,4*g*y(1)])
    plt.show()
    Tax2 = round(x(1)-NP,2)
    DWLS1D2 = (x(1) - np.interp(y(1),D2[::-1],Prices[::-1]))*(y(1)-e*y(1))/2
    DWLS1D3 = (np.interp(y(1),D3[::-1],Prices[::-1]) - x(1))*(g*y(1)-y(1))/2
    DWLS2D2 = (np.interp(NQD,D2[::-1],Prices[::-1]) - NP)*(e*NQD-NQD)/2
    DWLS2D3 = (np.interp(NQD,D3[::-1],Prices[::-1]) - NP)*(g*NQD-NQD)/2
    DWLSMCD2 = (SP - np.interp(SQD,D2[::-1],Prices[::-1]))*(SQD-e*SQD)/2
    DWLSMCD3 = (np.interp(SQD,D3[::-1],Prices[::-1]) - SP)*(g*SQD-SQD)/2
    if NQD > y(1):
        print("Subsidy: ", format(Tax2))
    else:
        print("Tax: ", format(-Tax2))
    print("DWLP1: ",format(DWLS1D2+DWLS1D3))
    print("DWLP2: ",format(DWLS2D2+DWLS2D3))
    print("DWLP3: ",format(DWLSMCD2+DWLS2D3))
interactive(lambda NQD,SQD: nqd2(NQD=NQD,SQD=SQD),
            SQD = BoundedFloatText(value=0, min = 0, max = 600*y(1), step=0.001),
            NQD = BoundedFloatText(value=0, min = 0, max = 600*y(1), step=0.001))
```
